# Demo

```python
import financial.data as fd
import financial.portfolio as fp
import financial.portfolios.statistics as stats

import os
```

```python
from financial.io.cache import NoUpdateStrategy
from financial.io.file.cache import FileCache

print(os.environ["DATA"])
print(os.environ["CACHE"])
ds = fd.CachedDataStore(path=os.environ["DATA"], cache=FileCache(cache_path=os.environ["CACHE"]+"/", update_strategy=NoUpdateStrategy()))

print(ds)
```

## Portfolio

```python
# Cartera de ejemplo (50% Tecnología, 50% Bienes de consumo básicos)
portfolio = fp.WeightedPortfolio("XLK+XLP", assets=[{'ticker':'XLK', 'weight': 0.5}, {'ticker':'XLP', 'weight': 0.5}])

# Benchmark (S&P 500)
benchmark = fp.BenchmarkPortfolio("^GSPC")

print(portfolio.get_assets())
```

## Portfolio returns

```python
start_date = '2013-12-31'
end_date = '2023-12-31'
```

```python
print("PORTFOLIO DAILY RETURNS")
portfolio_returns = portfolio.returns(ds, start_date, end_date)
print(portfolio_returns)

print("BENCHMARK DAILY RETURNS")
benchmark_returns = benchmark.returns(ds, start_date, end_date)
print(benchmark_returns)
```

```python
print("PORTFOLIO CUMULATIVE RETURNS")
portfolio_cumulative_returns = portfolio.cumulative_returns(ds, start_date, end_date)
print(portfolio_cumulative_returns)

print("BENCHMARK CUMULATIVE RETURNS")
benchmark_cumulative_returns = benchmark.cumulative_returns(ds, start_date, end_date)
print(benchmark_cumulative_returns)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import style

# Adjusting the size of matplotlib
import matplotlib as mpl

mpl.rc('figure', figsize=(16, 9))
mpl.__version__

# Adjusting the style of matplotlib
style.use('ggplot')

portfolio_cumulative_returns.index = pd.to_datetime(portfolio_cumulative_returns.index) # Index -> DateTimeIndex
benchmark_cumulative_returns.index = pd.to_datetime(benchmark_cumulative_returns.index) # Index -> DateTimeIndex

figure = plt.figure(figsize=(16, 9))
axis = figure.add_subplot(1, 1, 1)
axis.set_xlabel("Date")
axis.set_ylabel("Cumulative return")

axis.plot(100*portfolio_cumulative_returns, label="Portfolio")
axis.plot(100*benchmark_cumulative_returns, label="Benchmark")
axis.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), fancybox=True, ncol=2)

print(f"Cumulative returns, from {start_date} to {end_date}")
```

## Portfolio statistics

```python
def portfolio_statistics(returns, start_date, end_date):
    print(f"From {start_date} to {end_date}")
    print(f"Cumulative return     {100*stats.CumulativeReturn().get(returns):10.2f}%")
    print(f"Annualized return     {100*stats.AnnualizedReturn().get(returns,start_date,end_date):10.2f}%")
    print(f"Maximum drawdown      {100*stats.MaximumDrawdown().get(returns):10.2f}%")
    print(f"Annualized volatility {100*stats.AnnualizedVolatility().get(returns):10.2f}%")
    print(f"Sharpe ratio          {stats.SharpeRatio().get(returns):10.3f}")
    print(f"Sortino ratio         {stats.SortinoRatio().get(returns):10.3f}")

print("PORTFOLIO")
portfolio_statistics(portfolio_returns,start_date,end_date)

print("BENCHMARK")
portfolio_statistics(benchmark_returns,start_date,end_date)
```
