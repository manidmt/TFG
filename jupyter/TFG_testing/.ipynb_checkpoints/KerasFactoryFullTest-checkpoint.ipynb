```python
import tensorflow as tf
from keras import backend as K
import gc

def reset_gpu():
    K.clear_session()
    tf.keras.backend.clear_session()
    gc.collect()
```

```python
load_dotenv(dotenv_path=find_dotenv())

cache = os.environ["CACHE"]
model = os.environ["MODEL"]
print(model)
print(os.environ["DATA"])
datastore = fd.CachedDataStore(path=os.environ["DATA"], cache=FileCache(cache_path=cache+"/"))
print(datastore) 
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import style

# Adjusting the size of matplotlib
import matplotlib as mpl

mpl.rc('figure', figsize=(16, 9))
mpl.__version__
style.use('ggplot')
def plot_series(series: list[pd.Series], start_date: str=None, end_date: str=None):
    for i, serie in enumerate(series):
            linewidth = 1.0  # Grosor de línea predeterminado
            if i == 0:  # Si es la primera serie en la lista (índice 0)
                linewidth = 3.0  # Grosor de línea mayor
            serie[start_date:end_date].plot(linewidth=linewidth)
```

```python
ticker = "^GSPC"
start_date = "1990-01-01"
end_date = "2023-12-31"
```

```python
data = datastore.get_data(ticker, start_date, end_date)
target = data[90 + 20:]
metrics_results = {}
```

```python
def metrics(experiment, predictions):
    metrics_multiple = Experiment(experiment.name, predictions, target)
    r2 = r2_score(target, predictions)
    
    print("GLOBAL:")
    print(f"n={metrics_multiple.samples()} MSE={metrics_multiple.MSE():.4f} RMSE={metrics_multiple.RMSE():.4f} MAE={metrics_multiple.MAE():.4f} MAPE={metrics_multiple.MAPE():.4f} R² = {r2:.4f}")
    
    metrics = {
        "n": metrics_multiple.samples(),
        "MSE": metrics_multiple.MSE(),
        "RMSE": metrics_multiple.RMSE(),
        "MAE": metrics_multiple.MAE(),
        "MAPE": metrics_multiple.MAPE(),
        "R2": r2
    }
    metrics_results[experiment.name] = metrics
    if (len(experiment.train.samples())):
        print("TRAIN: ")
        results = experiment.train
        print(f"n={results.samples()} MSE={results.MSE():.4f} RMSE={results.RMSE():.4f} MAE={results.MAE():.4f} MAPE={results.MAPE():.4f}")
    
    if (len(experiment.test.samples())):
        print("TEST: ")
        results = experiment.test
        print(f"n={results.samples()} MSE={results.MSE():.4f} RMSE={results.RMSE():.4f} MAE={results.MAE():.4f} MAPE={results.MAPE():.4f}")
```

```python
import numpy as np
import tensorflow as tf
import time

# Simula las dimensiones de tus datos
n_samples = 5959
timesteps = 90
n_features = 1
batch_size = 32
epochs = 20 # Solo para un test rápido, puedes reducirlo si quieres

# Crea datos dummy (aleatorios) en memoria
print("Creando datos dummy...")
dummy_X_train = np.random.rand(n_samples, timesteps, n_features).astype(np.float32)
dummy_y_train = np.random.rand(n_samples, 1).astype(np.float32)
print("Datos dummy creados.")

# Define el modelo (copia la lógica de inicialización de RecurrentModel)
# Esto simula tu initialize_model
print("Construyendo modelo...")
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Input(shape=(timesteps, n_features)))
model.add(tf.keras.layers.LSTM(32, return_sequences=True))
model.add(tf.keras.layers.LSTM(16))
model.add(tf.keras.layers.Dense(1, activation='linear')) # Tu output activation es 'linear'

model.compile(
    loss="mean_squared_error",
    optimizer="adam",
    metrics=["mae"]
)
model.summary() # Para ver la arquitectura
print("Modelo construido.")

# Entrena el modelo con los datos dummy
print(f"Entrenando modelo con {n_samples} muestras dummy...")
start_time = time.time()
history = model.fit(
    dummy_X_train, dummy_y_train,
    epochs=epochs,
    batch_size=batch_size,
    validation_split=0.1
)
end_time = time.time()
print(f"Tiempo total de entrenamiento para {epochs} epochs: {end_time - start_time:.2f} segundos")
print(f"Tiempo promedio por epoch: {(end_time - start_time) / epochs:.2f} segundos")
```
